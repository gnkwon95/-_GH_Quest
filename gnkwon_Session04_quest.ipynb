```python
import matplotlib.pyplot as plt #matplotlib가 한글폰트를 지원 안해서 미리 처리해주는 과정
import platform                 #아마 요즘 matplotlib는 이런거 안해도 잘 될수도 있어요!

from matplotlib import font_manager, rc
plt.rcParams['axes.unicode_minus']= False

if platform.system() == 'Darwin': #맥os 사용자
    rc('font', family = 'AppleGothic')
elif platform.system() == 'Windows':#윈도우 사용자
    path = 'c:/Windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
```

```python
data_result =pd.merge(CCTV_Seoul, pop_Seoul, on = '지역구별')
```

```python
data_result.head()
```

```python
CCTV_reg = np.polyfit(data_result['인구수'], data_result['소계'], 1)
CCTV_reg

# add data_result that merges CCTV and pop
```

```python
f1 = np.poly1d(CCTV_reg)
fx = np.linspace(100000, 700000, 100)
```

```python
plt.figure(figsize=(6,6))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color ='g')
plt.xlabel('인구수')
plt.ylabel('소계')
plt.grid()
plt.show()
```

```python
data_result['오차'] = abs(f1(data_result['인구수']) - data_result['인구수'])
```

```python
data_result.head()
```

```python
data_result.sort_values(by='오차')
```

```python
x=data_result['오차']
y=data_result['노인비율']
colormap = data_result['오차']

plt.figure(figsize=(10,6))
plt.scatter(x,y, s=50, c = colormap, marker='>') 

for i in range(len(data_result)):
    plt.text(x[i]*1.02, y[i]*0.98, data_result['지역구별'][i], fontsize = 15) #plt.text(x축좌표, y축좌표, 메세지, 폰트사이즈)
    
plt.colorbar()
plt.show()
```

```python
x=data_result['오차']
y=data_result['고령자']
colormap = data_result['오차']

plt.figure(figsize=(10,6))
plt.scatter(x,y, s=50, c = colormap, marker='>') 

for i in range(len(data_result)):
    plt.text(x[i]*1.02, y[i]*0.98, data_result['지역구별'][i], fontsize = 15) #plt.text(x축좌표, y축좌표, 메세지, 폰트사이즈)
    
plt.colorbar()
plt.show()
```

```python
#인구수가 많을수록 오차가 커지고, 이 비율은 아주 선형적이다.  
#인구수가 많을수록 오차가 커지기에 고령자가 많으면 오차가 느는것처럼 보이지만 사실 노인비율로 보면 그 반대의 결과를 볼 수 있다
#이는 아마 노인이 많을수록 상대적 신생아의 비율이 적고, 이에 따라 인구수 예측의 정확성이 오르기 때문일 수 있다.
```
