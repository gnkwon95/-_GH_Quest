```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()

print(cancer.keys())
```

```python
X = cancer['data']
Y = cancer['target']

X, Y, cancer.DESCR
```

```python
import numpy as np
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
```

```python
X_radius = X[:, 0].reshape(-1,1)
Y = (Y==0).astype(np.int)

log_reg = LogisticRegression()
log_reg.fit(X_radius,Y)
```

```python
X_radius
```

```python
X_new = np.linspace(0, 20, 1000).reshape(-1,1)
y_prob = log_reg.predict_proba(X_new)
plt.plot(X_new, y_prob[:,1], 'g-', label='Breast cancer')
plt.plot(X_new, y_prob[:,0], 'b-', label='Not breast cancer')
```
